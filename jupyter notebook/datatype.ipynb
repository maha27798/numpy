Data Types in NumPy
NumPy has some extra data types, and refer to data types with one character, like i for integers, u for unsigned integers etc.
Below is a list of all data types in NumPy and the characters used to represent them.
i - integer
b - boolean
u - unsigned integer
f - float
c - complex float
m - timedelta
M - datetime
O - object
S - string
U - unicode string
V - fixed chunk of memory for other type ( void )

```python
import numpy as np
arr = np.array([1, 2, 3, 4])
print(arr.dtype)
```

```python
import numpy as np
arr = np.array([1.1, 2.0, 3.8, 4.9])
print(arr.dtype)
```

```python
import numpy as np
arr = np.array(['apple', 'banana', 'cherry'])
print(arr.dtype)
```

Creating Arrays With a Defined Data Type

```python
import numpy as np
arr = np.array(['apple', 'banana', 'cherry'], dtype="S")
print(arr)
print(arr.dtype)
```

```python
import numpy as np

arr = np.array([1, 2, 3, 4], dtype='i4')

print(arr)
print(arr.dtype)
```

```python
import numpy as np

arr = np.array([1.1, 2.3, 3.4, 4.7], dtype='f4')

print(arr)
print(arr.dtype)
```

Converting Data Type on Existing Arrays

```python
import numpy as np
arr=np.array([1,2,3,4,5,6])
newarray= arr.astype("f")
print(newarray, arr.dtype)
```

```python
import numpy as np
arr=np.array([1,0,0,0])
newarray= arr.astype("S")
print(newarray, arr.dtype)
```

```python
import numpy as np
arr=np.array([1.2,3.4,5.6])
newarray= arr.astype("i")
print(newarray, arr.dtype)
```
